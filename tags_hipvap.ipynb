```python
import re
template_variable = {
        "current": {
          "selected": False,
          "text": "t1",
          "value": "t1"
        },
        "hide": 0,
        "includeAll": False,
        "multi": False,
        "name": "Tags",
        "options": [
         
        ],
        "query": "t1, t2",
        "skipUrlSync": False,
        "type": "custom"
      }
def create_tag_template():
    tag_template = {
                  "selected": True,
                  "text": "tag",
                  "value": "tag"
                }
    return tag_template
def prepare_panel(board_name, tags, uuid_list, board_tag):
    file_name = 'panel_asset'
    if board_name== 'HT Coefficient Board':
        file_name = 'ht_coeff'
    with open(f'{file_name}.json', 'r') as f:
        panel_data_ = json.load(f)
    panel_data = panel_data_.copy()
    option_list = []
    if tags:
        for tag in tags:
            tag_template = create_tag_template()
            if tags.index(tag) == 0:
                tag_template["selected"] = True
            else:
                tag_template["selected"] = False
            tag_template["text"] = str(tag)
            tag_template["value"] = str(tag) 
            option_list.append(tag_template)
        current = {
              "selected": True,
              "text": tags[0],
              "value": tags[0]
            }
        panel_data["templating"]['list'][0] = template_variable
        panel_data["templating"]['list'][0]["current"] = current
        panel_data["templating"]['list'][0]['options'] = option_list    
        panel_data["templating"]['list'][0]["query"] =  ','.join(tags)
    panel_data["tags"] = [board_tag]
    uuid = ''.join(random.choices(string.ascii_letters + string.digits, k=10))
    while uuid in uuid_list:
        uuid = ''.join(random.choices(string.ascii_letters + string.digits, k=10))
        
        
    panel_data["version"] += 1
    panel_data["uid"] = uuid
    panel_data["time"]["from"] = "now-24h"
    panel_data["time"]["to"] = "now"
    panel_data['title'] = board_name
    with open(f'{board_name}.json', 'w', encoding='utf-8') as f:
        json.dump(panel_data, f, ensure_ascii=False, indent=4)
    return uuid
```

```python
def prepare_md_text(header_text, uuid_dict_for_md_text):
    title_text = f"## {header_text}  \r\n"
    text_list = []
    endtext = "from=${__from}&to=${__to})  \r\n"
    for title, uuid in uuid_dict_for_md_text.items():
        b = re.sub('[^A-Za-z0-9]+', '-', title)
        if b[-1]=='-':
            b = b[:-1]
        text_for_asset = f"[{title}](/d/{uuid}/{b.lower()}?&"
        text_for_asset = text_for_asset + endtext

        text_list.append(text_for_asset)
    whole_text = ''.join(text_list)

    md_text = title_text + whole_text
    return md_text
```

```python
# uuid_list = []
# with open('assets_board.json', 'r') as f:
#         panel_data = json.load(f)

def prepare_list_board(md_text, title, board_tag):
    file_name = 'assets_board'
    if title== 'HT Coefficient Board':
        file_name = 'ht_coeff'
    with open(f'{file_name}.json', 'r') as f:
        panel_data_ = json.load(f)
    panel_data = panel_data_.copy() 
    options = {'content': md_text, "mode": "markdown"}
    panel_data['panels'][0]['options'] = options
    uuid = ''.join(random.choices(string.ascii_letters + string.digits, k=10))
    while uuid in uuid_list:
            uuid = ''.join(random.choices(string.ascii_letters + string.digits, k=10))
    panel_data['uid'] = uuid
    panel_data["tags"] = [board_tag]
    panel_data['title'] = title
    with open(f'{title}.json', 'w', encoding='utf-8') as f:
            json.dump(panel_data, f, ensure_ascii=False, indent=4)
    return uuid
```

```python
def create_link_template(title, uid):
    title_grafana = re.sub('[^A-Za-z0-9]+', '-', title)
    if title_grafana[-1]=='-':
        title_grafana = title_grafana[:-1] 
    link_template = {
      "asDropdown": False,
      "icon": "dashboard",
      "includeVars": False,
      "keepTime": False,
      "tags": [],
      "targetBlank": False,
      "title": title,
      "tooltip": "",
      "type": "link",
      "url": f"/d/{uid}/{title_grafana}?orgId=1"
    }
    return link_template
```
